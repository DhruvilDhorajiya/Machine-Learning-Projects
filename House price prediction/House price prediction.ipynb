### House price prediction ###

Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
data = pd.read_csv("boston.csv")
```

```python
house_price_dataframe = pd.DataFrame(data)
```

```python
house_price_dataframe.head(5)
```

```python
house_price_dataframe.shape
```

```python
#check for missing values
house_price_dataframe.isnull().sum()
```

```python
#Statatical measures of dataset
house_price_dataframe.describe()
```

```python
#Corelating features
correlation = house_price_dataframe.corr()
```

```python
#Constructing heatmap to understand correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
#Spliting dataset into data and target(Value)

X = house_price_dataframe.drop(["PRICE"],axis=1)
Y = house_price_dataframe["PRICE"]
```

```python
print(X)
print(Y)
```

```python
#splitting the dataset into training and testing
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Training the model
#XGBregressor

#loading the model
model = XGBRegressor()
```

```python
#Training the model
model.fit(X_train,Y_train)
```

```python
#Evaluation of model
#predecting on training data
train_data_prediction = model.predict(X_train)
```

```python
print(train_data_prediction)
```

```python
#R square error
score_1 = metrics.r2_score(Y_train,train_data_prediction)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,train_data_prediction)


print("R-Square erroe: ",score_1)
print("Mean absolute_error: ",score_2)
```

```python
#Visualizing the actual price and predicted price
plt.scatter(Y_train,train_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted Prices")
plt.title("Actual V/S Predicted on training data")
plt.show()
```

```python
#prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
#Accuracy on test data

#R-square error
score_1 = metrics.r2_score(Y_test,test_data_prediction)


#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)

print("Test data R-square error: ",score_1)
print("Test data Mean absolute error: ",score_2)
```
